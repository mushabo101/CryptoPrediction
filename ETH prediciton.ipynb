```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('ETH DATA SET.csv')
```

```python
from datetime import datetime
df['timestamp'] = [datetime.fromtimestamp(X) for X in df['timestamp']]
```

```python
df.head()
```

```python
df.drop(df.loc[df['high'] == 'undefined'].index, inplace=True)
```

```python
df = df.sort_values('timestamp',ascending=True)
df = df.reset_index()
```

```python
df
```

```python
projection = 14
df['prediction'] = df[['close']].shift(-projection)
```

```python
#Create the independant data set(X)
#Convert the dataFrame to an array
X = np.array(df[['close']])
#remove the last 14 rows of data
X = X[:-projection]
print(X)
```

```python
#Create the dependat data set (y)
#convert the dataframe to array

y = df['prediction'].values
#remove the last 14 rows of data
y = y[:-projection]
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size = 0.20)
```

```python
#Create and train the linear Regression Model
linReg = LinearRegression()
linReg.fit(X_train,y_train)
```

```python
linReg_confidence = linReg.score(X_test,y_test)
print(linReg_confidence)
```

```python
#create and set x_projection equal to the last 14 rows of the original data set
x_projection = np.array(df[['close']])[-projection:]
print(x_projection)
```

```python
linReg_prediction = linReg.predict(x_projection)
print(linReg_prediction)
```
